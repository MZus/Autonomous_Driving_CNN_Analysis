```python
i = 0
indicies = [] #main incidies list

while i < len(df)-1:
    
    #temp indicies for each picture 
    temp_indicies = []
    
     #getting the first occurance of the file
    name = df['filename'][i]
    
    #adding the first occurance
    temp_indicies.append(i)  
    j = i + 1 #iterating
    
    while name == df['filename'][j] and i < len(df)-3: #second while loop to add the other indicies 
        temp_indicies.append(j) #adding the next indicie(s) 
        j += 1 #iterating
    indicies.append(temp_indicies) #adding the temp_indicies list to the list of indicies to get a 2D list
    i = j #updating j to i to ensure no duplicaets are added



```

```python
print(indicies[:10]) #checking our list is how we wanted it
```

```python
for image_list in indicies:
    label_list = []
    for i in image_list:
        label_list.append("1")
  #  print(i)
```

```python

#setting up blank lists
both_list = []
car_list = []
pedestrian_list = []
neither_list = []

#iterating through our 2D list of indicies
for image_list in indicies:
    
    #making a temp list
    label_list = []  
    
    for i in image_list:
    
        #adding the classes for the picture in the list one by one
        label_list.append(df["class"][i]) 
        
    #checking to see which group each index in the temp list (label_list) is in
    if "car" in label_list and  "pedestrian" in label_list: 
        both_list.append(i)
    elif "car" in label_list: 
        car_list.append(i)
    elif "pedestrian" in label_list:
        pedestrian_list.append(i)
    else:
        neither_list.append(i)
```

```python
both_images = []
car_images = []
pedestrian_images = []
neither_images = []

IMAGE_SIZE = 100

for i in both_list:
    name = df['filename'][i]
    try:
        image = Image.open(pic_dir + name)
        image = image.resize((IMAGE_SIZE,IMAGE_SIZE))
        image = ImageOps.grayscale(image)
        
        both_images.append(np.array(image))
        
        #temp_image = np.asarray(image)
    except Exception as e:
        pass #to pass the error and not print anything out
    
    image.close()

for i in car_list:
    name = df['filename'][i]
    try:
        image = Image.open(pic_dir + name)
        image = ImageOps.grayscale(image)
        image = image.resize((IMAGE_SIZE,IMAGE_SIZE))
        car_images.append(np.array(image))
       # temp_image = np.asarray(image)
    except Exception as e:
        pass
    
    image.close()
       
for i in pedestrian_list:
    name = df['filename'][i]
    try:
        image = Image.open(pic_dir + name)
        image = ImageOps.grayscale(image)
        image = image.resize((IMAGE_SIZE,IMAGE_SIZE))
        pedestrian_images.append(np.array(image))
        #temp_image = np.asarray(image)
    except Exception as e:
        pass
    
    image.close()
    
for i in neither_list:
    name = df['filename'][i]
    try:
        image = Image.open(pic_dir + name)
        image = ImageOps.grayscale(image)
        image = image.resize((IMAGE_SIZE,IMAGE_SIZE))
        neither_images.append(np.array(image))
         #temp_image = np.asarray(image)
    except Exception as e:
        pass
        

    image.close()


    


    
```
